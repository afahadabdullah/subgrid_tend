```python
import xarray as xr
import numpy as np
import os
from glob import glob
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
dir='/nobackupp27/dmenemen/public/geos_ecco/c1440_llc2160/holding/'
odir='/nobackupp27/afahad/project/subgrid_tenden/data/'
os.chdir(odir)
```

```python
dfiles=sorted(glob('dT_*.nc4'))
files=sorted(glob('C90T_*.nc4'))
files[:5]
```

```python
dir='/nobackupp27/dmenemen/public/geos_ecco/c1440_llc2160/holding/inst_01hr_3d_T_Mv/'
```

```python
# create Hres dprdt

for i in tqdm(range(len(files))):
            ds1=xr.open_dataset(files[i])
            ds2=xr.open_dataset(files[i+1])
            dT=ds2.T.data-ds1.T
            HdT=xr.open_dataset('dT'+files[i][4:])
            sub=HdT.dT-dT
            Ttrain=xr.merge([sub.rename('subT'),dT.rename('dT'),ds1.T])
            cc=xr.open_mfdataset(dir+files[i][5:])
            #print(cc)
            Ttrain.coords.update(cc.coords)
            Ttrain.to_netcdf(odir+'Ttrain_'+files[i])
```

```python
plt.imshow(Ttrain.subT[0,-1,1,:,:],cmap='bwr')
plt.colorbar()
```
